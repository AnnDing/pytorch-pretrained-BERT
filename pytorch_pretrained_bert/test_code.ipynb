```python
import argparse
import logging
from tqdm import trange

import torch
import torch.nn.functional as F
import numpy as np

from pytorch_pretrained_bert import GPT2LMHeadModel
from tokenization_gpt2 import GPT2Tokenizer
```

```python
np.random.seed(0)
torch.random.manual_seed(0)
torch.cuda.manual_seed(0)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

enc = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.to(device)
print(model)
model.eval()
```

```python
raw_text = 'I bought a'
context_tokens = enc.encode(raw_text.strip())
context = torch.tensor(context_tokens, device=device, dtype=torch.long).unsqueeze(0).repeat(1, 1)
past = None
```

```python
logits, past = model(context, past=past)
logits = logits[:, -1, :] / 1.0
values, indexes = torch.topk(logits, 100)
prob = F.softmax(values, dim=-1)

topk_tokens = []
for index in indexes.tolist()[0]:
    topk_tokens.append(enc.decode([index]))

    for index, token, p in zip(indexes.tolist()[0], topk_tokens, prob.tolist()[0]):
        print('{} : {} , {}'.format(index, token, p))
```
